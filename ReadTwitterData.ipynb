```python
file_list = glob.glob('data*.json')

dfs = [] # an empty list to store the data frames
for file in file_list:
    data = pd.read_json(file) # read data frame from json file
    dfs.append(data) # append the data frame to the list

temp = pd.concat(dfs, ignore_index=True) # concatenate all the data frames in the list
```

```python
len(temp)
```

```python
temp = temp.sort_values(by=['created_at'])
```

```python
temp.head()
```

```python
import ast
from pandas.io.json import json_normalize
geo = pd.json_normalize(temp['geo'])
public_metrics = pd.json_normalize(temp['public_metrics'])
temp.drop(['geo','public_metrics'],axis=1,inplace=True)
```

```python
#creating new concatenated dataframe
temp.reset_index(drop=True, inplace=True)
geo.reset_index(drop=True, inplace=True)
public_metrics.reset_index(drop=True, inplace=True)
df = pd.concat([temp,geo,public_metrics],axis=1)
```
