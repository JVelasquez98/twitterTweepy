```python
import tweepy as tw
import os
import pandas as pd
import datetime
```

```python
twitter_keys = {
        'consumer_key':        '',
        'consumer_secret':     '',
        'access_token_key':    '',
        'access_token_secret': ''
    }
    
```

```python
#Setup access to API
auth = tw.OAuthHandler(twitter_keys['consumer_key'], twitter_keys['consumer_secret'])
auth.set_access_token(twitter_keys['access_token_key'], twitter_keys['access_token_secret'])

#api = tw.API(auth)
api = tw.API(auth, wait_on_rate_limit=True)
```

```python

#Make call on home timeline, print each tweets text
public_tweets = api.home_timeline()
for tweet in public_tweets:
    print(tweet.text)
    
```

```python
def get_tweets(query, 
#geo, 
count, page, start, end):
    '''
    Main function to fetch tweets and parse them.
    '''
    # empty list to store parsed tweets
    tweets = []
 
    try:
        # call twitter api to fetch tweets

        for tweet in tw.Cursor(api.search, q=query, 
        #geocode=geo, 
        until=end).items(count):

            # appending parsed tweet to tweets list
            if tweet.retweet_count > 0:
            # if tweet has retweets, ensure that it is appended only once
                if tweet not in tweets:
                    tweets.append(tweet)
            else:
                tweets.append(tweet)
 
            # return parsed tweets
        return tweets
 
    except tw.TweepError as e:
        # print error (if any)
        print("Error : " + str(e))

#need cordinates and milage around to try and specifiy tweets

places = api.geo_search(query="Congo", granularity="country")
place_id = places[1].id

geo='0.2280,15.8277,25km'

#place should indicate the id of the area "Congo - Brazaville"
tweets = get_tweets(
    query = "place:%s" %place_id 
    #query = "", geo = geo
    , count=3000, page = 5, start = datetime.date.today() - datetime.timedelta(days=30),end=datetime.date.today())
```

```python
json_data = [r._json for r in tweets]

df = pd.json_normalize(json_data)
df
```
